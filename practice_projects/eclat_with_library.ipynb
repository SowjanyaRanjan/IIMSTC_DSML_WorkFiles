```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(
    r'C:\Users\sowja\OneDrive\Desktop\practice_projects\Market_Basket_Optimisation.csv',
    header=None
)

dataset.head()
```

```python
transactions = []

for i in range(0, 7501):
    transactions.append([str(dataset.values[i, j]) for j in range(0, 20)])
df = pd.DataFrame(transactions)
```

```python
from pyECLAT import ECLAT

eclat = ECLAT(data=df)
indexes, support = eclat.fit(min_support=0.03, min_combination=2, max_combination=2)
```

```python
print("Frequent Itemsets:")
print(indexes)
```

```python
result_df = pd.DataFrame({
    "Itemset": indexes.keys(),
    "Support": support.values()
})

result_df.sort_values(by="Support", ascending=False)
```
